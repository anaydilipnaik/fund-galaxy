```python
investors = pd.read_csv('data/investors.csv')
investments = pd.read_csv(r'data/investments.csv')
partners = pd.read_csv(r'data/investment_partners.csv')
org = pd.read_csv(r'data/organizations.csv')
```

```python
df = pd.DataFrame()
df['uuid'] = partners['investor_uuid']
df['name'] = partners['investor_name']
df['type'] = investments['investor_type']
df['partners'] = partners['partner_name']
df['tot_funding'] = org['total_funding']
df['address'] = investors['city']+','+investors['region']
df['linkedin'] =  investors['linkedin_url']
df['domain'] = partners['cb_url']
df['description'] = partners['name']
df['invested_companies'] = partners['funding_round_name']
df['investment_type'] = investors['investor_types']
```

```python
df.isnull().sum()
```

```python
df['tot_funding'] = df['tot_funding'].fillna(df['tot_funding'].mean())
df['address'] = df['address'].fillna('San Fransico,California')
df['linkedin'] = df['linkedin'].fillna('Unknown')
df['investment_type'] = df['investment_type'].fillna('equity firm,venture capital')
```

```python
df = df.drop(columns='adress')
```
