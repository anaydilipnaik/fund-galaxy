```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ast import literal_eval
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
org1 = pd.read_csv(r'data/organizations.csv')
people = pd.read_csv(r'data/people.csv')
founder = pd.read_csv(r'data/people_descriptions.csv')
```

```python
org1.info()
```

```python
org1['uuid'][0]
```

```python
cols = ['uuid','name','short_description','domain','total_funding','num_funding_rounds','total_funding_currency_code','founded_on','employee_count','email','phone','address','facebook_url','linkedin_url','twitter_url']
```

```python
df = org1[cols].copy()
```

```python
df.isnull().sum()
```

```python
df['total_funding_currency_code'].value_counts()
```

```python
df['total_funding'] = df['total_funding'].fillna(df['total_funding'].mean())
df['domain'] = df['domain'].fillna('unknown')
df['num_funding_rounds'] = df['num_funding_rounds'].fillna(2)
df['total_funding_currency_code'] = df['total_funding_currency_code'].fillna('USD')
df['founded_on'] = df['founded_on'].fillna(2000-6-27)
df['email'] = df['email'].fillna('info@crunchbase.com')
df['phone'] = df['phone'].fillna('423-111-2929')
df['facebook_url'] = df['facebook_url'].fillna('https://www.facebook.com/cruchbase')
df['linkedin_url'] = df['linkedin_url'].fillna('https://www.linkedin.com/cruchbase')
df['twitter_url'] = df['twitter_url'].fillna('https://www.twitter.com/cruchbase')
df['people'] = people['name']
df['founder'] = founder['name']
df['address'] = df['address'].fillna('100 Walnut Avenue Suite')
df['category'] = org1['category_list']
df['category_grp_list'] = org1['category_groups_list']
df = df.rename(columns = {'short_description':'description'})
df['category'] = df['category'].fillna('Education, Health')
df['category_grp_list'] = df['category_grp_list'].fillna('Events, Media')
# df['category'] = df['category'].str.split(',')
# df['category_grp_list'] = df['category_grp_list'].str.split(',')
```
